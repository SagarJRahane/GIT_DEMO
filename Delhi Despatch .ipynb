```python
from calendar import monthrange
month=monthrange(2022,9)
month[-1]
```

```python
def data_format(last):
    urls=[]
    for i in range (1,last+1):
        if i<10:
            urls.append(f"http://www.delhisldc.org/Loaddata.aspx?mode=0{i}/09/2022")
        else:
            urls.append(f"http://www.delhisldc.org/Loaddata.aspx?mode={i}/09/2022")
    return urls

S=data_format(month[-1])
S
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="http://www.delhisldc.org/Loaddata.aspx?mode=17/09/2022"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
all_data=soup.find_all('tbody')[2]
```

```python
all_data=soup.find_all('tbody')[2]
all_data
```

```python
table=all_data.find_all('tr')
data=[d.find_all('td') for d in table]
print(data)
result=[[j.get_text() for j in i] for i in data]

# result
# for i in data:
#     p=[]
#     for j in i:
#         p.append (j.get_text())
#     print(p)
#     break
    
```

```python
d1={k:[] for k in result[0]}
for t in result[1:]:
    d1['TIMESLOT'].append (t[0])
    d1['DELHI'].append (t[1])
    d1['BRPL'].append (t[2])
    d1['BYPL'].append (t[3])
    d1['NDPL'].append (t[4])
    d1['NDMC'].append (t[5])
    d1['MES'].append (t[6])
df=pd.DataFrame(d1)
df
```

```python
df.to_csv(f'./Delhi Despatch Spet 2022/delhi_despatch_all.csv')
```
